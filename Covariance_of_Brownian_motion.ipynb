```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Simulate Brownian motion multiple times
matrix=np.zeros((10000,90))# create a matrix, 1000 rows, and 90 columns
for i in range(0,10000):
    brw1= np.cumsum(np.random.randn(10000))
    matrix[i,:] = brw1[10:100] * brw1[100]# multiply 90 entries from the first simulations with the 100th entry from each
    # simulation
# I need to average across rows to get the expecation of the product
```

```python
print(matrix.mean(axis=0))
```

```python
plt.plot(range(10,100),matrix.mean(axis=0))
plt.xlabel('Observation index')
plt.ylabel("Covariance between the x-axis obs. and 100th obs.")
plt.show()
```
